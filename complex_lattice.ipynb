```python
"""complex_lattice.ipynb"""

# Cell 1 

from __future__ import annotations

import typing
import matplotlib.pyplot as plt
import numpy as np

if typing.TYPE_CHECKING:
    from matplotlib.axes import Axes
    from numpy.typing import NDArray

%matplotlib widget

def complex_f(z: complex) -> complex:
    """ This defines the complex function """
    return z**2 + 1j * z + 1

def plot(ax: Axes) -> None:
    """ Plots the scatter plot of the function """

    # Defining the bounds of the plot
    real_bounds: NDArray[np.float_] = np.linspace(-4, 4, 1000)
    imaginary_bounds: NDArray[np.float_] = np.linspace(0, 2, 1000)

    # Meshgrid obtained from David Beirsach's code
    real, imaginary = np.meshgrid(real_bounds, imaginary_bounds)
    # Meshgrid creates a 2d grid of coordinates

    # Storing real and imaginary into another array once passed through the complex function
    z = complex_f(real + 1j * imaginary)


    # Plotting the Function as scatter points
    ax.scatter(z.real, z.imag, s=1)

    # Customizing the Plot
    ax.set_title(r"$f(z) = z^2 + iz + 1$")
    ax.set_ylabel('Imaginary')
    ax.set_xlabel('Real')
    ax.grid()
    ax.axis('equal')


def main() -> None:
    fig, ax = plt.subplots()
    plot(ax)
    plt.show()

main()
```

```python
# Cell 2
# Used David's Code for this solution

# Intitializing to zero
num_points = 0

# Iterating 
for x in range(-50, 50):
    for y in range(-50,50):
    # Conditional 
        if abs(x**2 - y**2 - y + 1) <= 10 and abs(2 * x * y + x) <= 10:
            num_points += 1

print(f"There are {num_points:,} lattice points.")
```
